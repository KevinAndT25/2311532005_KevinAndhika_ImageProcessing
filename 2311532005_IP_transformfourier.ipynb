```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data citra 3x3
citra = np.array([
    [148, 164, 174],
    [144, 159, 167],
    [128, 147, 157]
])

print("Citra asli (3x3):")
print(citra)
```

```python
# Melakukan DFT 2D menggunakan FFT dari NumPy
dfd = np.fft.fft2(citra)

print("Hasil DFT (bilangan kompleks):")
print(dfd)
```

```python
# Menampilkan magnitude (spektrum)
magnitude = np.abs(dfd)
print("Magnitude (spektrum):")
print(magnitude)
```

```python
# Menampilkan fase
phase = np.angle(dfd)
print("Fase (radian):")
print(phase)
```

```python
# Menampilkan komponen real dan imajiner
print("Bagian Real:")
print(dfd.real)
print()
print("Bagian Imajiner:")
print(dfd.imag)
```
